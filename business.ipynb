```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 21)

with open(r"C:\Users\Abdulaziz\DataSets\Sample - Superstore.csv") as f:
    print(f)
```

```python
df = pd.read_csv(r"C:\Users\Abdulaziz\DataSets\Sample - Superstore.csv", encoding='cp1252')
df.head()
```

```python
df.info()
```

```python
df_drops=['Order ID',
          'Country']
df.drop(df_drops, axis= 1)

df.set_index('Row ID')
```

```python
df.iloc[0:10]
```

```python
df.isnull().sum()
```

```python
dt_column = df.columns[1:3]
df[dt_column] = df[dt_column].apply(pd.to_datetime, errors= 'coerce')
df.info()
```

```python
sizes = df.groupby('Sub-Category')['Sales'].count().sort_values(ascending = False).head(10)
sizes
```

```python
labels = df['Sub-Category'].value_counts().index.tolist()[0:10]

fig1, pi1 = plt.subplots(figsize = (10,10))
pi1.pie(sizes, labels = labels, autopct = '%1.1f%%',textprops = {'fontsize': 14},
        shadow = True, radius = 4, startangle = 90)
pi1.axis('equal')
plt.show()
```

```python
profit_t10=plt.figure(figsize=(10,5))
plt.title('Top 10 cities based on profit')
plt.xlabel('City')
plt.ylabel('Sales')
df.groupby('City')['Profit'].sum().sort_values(ascending=False)[0:10].plot(kind='bar',color='Blue',ls='solid',edgecolor='White')
```

```python
custbase_t10=plt.figure(figsize=(10,8))
plt.title('Top 10 cities based on customer base')
plt.xlabel('Customer Base')
plt.ylabel('City')
df.groupby('City')['Customer ID'].count().sort_values(ascending=False)[0:10].plot(kind='barh',color='cyan',ls='solid',edgecolor='White')
```
